```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1 to read the csv
df=pd.read_csv(r'zomato.csv',encoding="ISO-8859-1")
df
```

```python
#2 Display number of columns,info,describe,shape
df.info()
print(df.describe())
df.shape
```

```python
#3 check how many null values are there
null_count=df.isnull().sum().sum()
print('Number of null values are:',null_count)
```

```python
#4 Read another excel file
df1=pd.read_excel('Country-Code.xlsx')
df1
```

```python
#5 merge both with country code column
df=pd.merge(df,df1,on="Country Code")
df
```

```python
#6 display final df shape,columns
df.info()
df.shape
```

```python
#7 plot pie chart
import numpy as np
plt.pie(df.groupby('Country').count()['Restaurant ID'], labels = np.sort(df.Country.unique()),autopct='%1.1f%%', textprops={'fontsize': 5})
```

```python
#8 plot pie chart for highest 3
country_name=df.Country.value_counts().index
country_val=df.Country.value_counts().values
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

```python
#9 groupby Agg rating,Rating color,Rating Text
result1=df.groupby(['Aggregate rating','Rating color','Rating text']).size()
result1
```

```python
#10 Reset the index name as Rating Counts 
result2=df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Counts'})
result2
```

```python
#11 Bar plot Agg Rating vs Rating Count
import seaborn as sns
sns.barplot(x='Aggregate rating',y='Rating Counts',data=result2)
plt.show()
```

```python
#12 bar plot with hue as Rating Color
sns.barplot(x='Aggregate rating',y='Rating Counts',hue='Rating color',data=result2)
plt.show()
```

```python
#13 barplot withe pallete
sns.barplot(x='Aggregate rating',y='Rating Counts',hue='Rating color',data=result2,palette=['white','red','orange','yellow','green','darkgreen'],edgecolor='black');
```

```python
#14 count plot with Rating Color vs Rating Counts
sns.countplot(x='Rating color',data=result2,palette=['white','red','orange','yellow','green','darkgreen'],edgecolor='black')
```

```python
#15 Group by country where Rating Color is white
result3=df[df['Rating color']=='White'].groupby(['Country']).size()
result3
```

```python
#16 Rename
result3=df[df['Rating color']=='White'].groupby(['Country']).size().reset_index().rename(columns={0:'Country Counts'})
result3
```

```python
#17 which currency is used by which country
result4=df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
result4[['Country','Currency']]
```

```python
#18 which country has online delivery option
result5=df[df['Has Online delivery']=='Yes'].Country.value_counts().reset_index()
result5['index']
```

```python
#19 Create pie chart for 5 top cities
city_values=df.City.value_counts().values
city_labels=df.City.value_counts().index
plt.pie(city_values[:5],labels=city_labels[:5], autopct='%1.2f%%');
```

```python
#20 Find top 10 cuisines
result5=df[['Cuisines']].groupby(['Cuisines']).size().reset_index().rename(columns={0:'Counts'})
print('Top 10 cuisines are:')
result5.sort_values(['Counts'],ascending=False)[:10]
```
